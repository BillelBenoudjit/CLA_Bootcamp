We may have these missing values because the dataset was corrupted during a transfer and some data went missing (Missing Completely at Random (MCAR)).

```python
#The list of columns with missings values within the dataset
df.loc[:, df.isnull().any()].columns
```

```python
#The mean, median, quartiles of each numerical variable within the dataset
describe = df.describe()
describe
```

```python
#Spotting the possible outliers in the dataset and get rid of them
df_zscore = (df - df.mean()) / df.std()
df = df[df_zscore < 3]
df
```
